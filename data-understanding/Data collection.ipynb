```python
import pandas as pd
from tqdm import tqdm
import time
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 30
```

```python
def get_data_from_urls(urls: list, encoding:str = 'utf-8') -> pd.DataFrame:
    """
    Get data from a list of urls.
    Params:
    - urls (list): list of urls to get data from.
    - encoding (str): encoding of the data, default is 'utf-8'.
    Returns:
    - pd.DataFrame: concatenated data from all urls.
    """
    data = []
    for url in tqdm(urls):
        d = pd.read_csv(url, encoding=encoding, low_memory=False)
        data.append(d)
        time.sleep(1.2)

    return pd.concat(data)
```

# Play by play

```python
pre_url = 'https://github.com/nflverse/nflverse-data/releases/download/pbp/play_by_play_'
end_url = '.csv'
urls = []

for i in range(1999, 2024):
    urls.append( pre_url + str(i) + end_url)
```

```python
pbp_df = get_data_from_urls(urls)
pbp_df.head()
```

```python
pbp_df.to_csv('pbp_1999_to_2023.csv', index=False)
```

# Player Stats (def, off, kicking)

```python
pre_url = 'https://github.com/nflverse/nflverse-data/releases/download/player_stats/player_stats_'
end_url = '.csv'
urls_stats = []
urls_stats_def = []
urls_stats_kick = []


for i in range(1999, 2024):
    urls_stats.append(pre_url + str(i) + end_url)
    urls_stats_def.append(pre_url + 'def_' + str(i) + end_url)
    urls_stats_kick.append(pre_url + 'kicking_' + str(i) + end_url)
```

```python
player_stats_df = get_data_from_urls(urls_stats)
player_stats_def_df = get_data_from_urls(urls_stats_def)
player_stats_kick_df = get_data_from_urls(urls_stats_kick)

display(player_stats_df.head())
display(player_stats_def_df.head())
display(player_stats_kick_df.head())
```

```python
player_stats_df.to_csv('player_stats_1999_to_2023.csv', index=False)
player_stats_def_df.to_csv('player_stats_def_1999_to_2023.csv', index=False)
player_stats_kick_df.to_csv('player_stats_kick_1999_to_2023.csv', index=False)
```
