```python
import pandas as pd
from ydata_profiling import ProfileReport
from tqdm import tqdm
```

# Play by play

```python
pbp_df = pd.read_csv('pbp_1999_to_2023.csv', low_memory=False)
pbp_df.head()
```

```python
# remove time due to errors
cols_remove_error = ['game_date', 'time', 'start_time', 'time_of_day', 'play_clock', 'drive_real_start_time', 'end_clock_time', 'drive_time_of_possession', 'drive_game_clock_start', 'drive_game_clock_end']
pbp_df.drop(cols_remove_error, axis=1, inplace=True)
profile = ProfileReport(pbp_df.sample(frac=0.1), title="Profiling pbp")
profile.to_file("Profiling pbp.html")
```
